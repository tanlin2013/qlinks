```python
A, B = list(nx.bipartite.sets(g))   # unpack the two node sets
n = g.number_of_nodes()

# Keep the node ordering consistent with the adjacency matrix:
nodes = list(g.nodes())  # the order used by nx.adjacency_matrix(g)

# Build diagonal values: +1 for nodes in A, -1 for nodes in B
vals = np.array([1 if node in A else -1 for node in nodes], dtype=float)

# Construct sparse diagonal matrix
c_opt = sp.diags(vals, format='csr')
c_opt.toarray()
```

```python
# vec = aut.type_3a_scars(8, fill_zeros=True)[0].evec[:,0]
#
# vecp = vec - c_opt @ vec
#
#
# for i, v in aut.type_1_scars((8, 'B'), fill_zeros=True)[1].evec.T:
#     print(np.abs(vecp.T @ v).item()**2)
#
#
def fidelity_matrix(M, N, normalize=False, tol=1e-15):
    """
    M, N: shape (d, k_M) and (d, k_N) with vectors as columns.
    Returns F of shape (k_M, k_N) with F[i,j] = |<M_i | N_j>|^2.
    """
    M = np.asarray(M, dtype=complex)
    N = np.asarray(N, dtype=complex)

    if normalize:
        # L2-normalize columns; avoid division by ~0
        M_norms = np.linalg.norm(M, axis=0)
        N_norms = np.linalg.norm(N, axis=0)
        M = M / np.maximum(M_norms, tol)
        N = N / np.maximum(N_norms, tol)

    overlaps = M.conj().T @ N          # shape (k_M, k_N)
    F = np.abs(overlaps)**2
    return F


vec = aut.type_3a_scars(8, fill_zeros=True)[0].evec


fidelity_matrix(
    c_opt @ (vec - c_opt @ vec),
    aut.type_1_scars((8, 'B'), fill_zeros=True)[0].evec
)
```

```python
# --- helpers ---
def set_dynamic_cycle(ax, n):
    colors = plt.cm.tab10_r(np.linspace(0, 1, n))
    # Repeating style lists to length n
    markers    = ["*", "x", "v", "^", "o", "d"]
    linestyles = ["-", "--", "-.", ":", "-", "--"]
    # Trim or repeat to match n
    markers    = (markers * ((n // len(markers)) + 1))[:n]
    linestyles = (linestyles * ((n // len(linestyles)) + 1))[:n]
    ax.set_prop_cycle(color=colors, marker=markers, linestyle=linestyles)

def pidx(x, y, Ly):  # plaquette index per your convention
    return y * Ly + x

def fidelity_vs_time(H, psi0, start, stop, num):
    # expm_multiply returns an iterator of state vectors at evenly spaced times
    evol_states = expm_multiply(-1j * H, psi0, start=start, stop=stop, num=num)
    fidel = np.array([np.abs(es.T @ psi0).item()**2 for es in evol_states])
    # rate_func = np.array([-np.log(es.T @ psi0).item() for es in evol_states])
    # entropy = [model.entropy(evol_state.flatten(), 1, 0) for evol_state in evol_states]
    return fidel

def build_model_and_H(coup_j, coup_rk, model_shape, basis, site1=None, site2=None, magni=None):
    mdl = QuantumLinkModel(coup_j, coup_rk, model_shape, basis)
    if magni is not None:
        p_corr = mdl.plaquette_corr_opt(site1, site2)
        return mdl.hamiltonian + magni * p_corr
    else:
        return mdl.hamiltonian


# scar_state = np.zeros((basis.n_states, 1))
# scar_state[np.array([5, 15, 74, 77]), 0] = np.array([1, -1, -1, 1])
# scar_state /= np.linalg.norm(scar_state)
scar_state = evecs[:, 56]


coup_j_base = coup_j * np.ones((np.prod(model.shape), 1), dtype=complex)
coup_rk_base = coup_rk * np.ones((np.prod(model.shape), 1), dtype=np.complex128)
start, stop, num = (0, 400, 150)
tgrid = np.linspace(start, stop, num)


fig, ax = plt.subplots(figsize=(12, 8), facecolor="white")
set_dynamic_cycle(ax, 8)

for target_plaquette1, target_plaquette2, pert_value in product(
    [(1, 1)],
    [
        (0, 1), (1, 0), (1, 2), (2, 1),
        # (0, 0), (2, 2), (0, 2), (2, 0),
        # (1, 3), (3, 1),
        # (3, 3),
        # (0, 3), (3, 0), (2, 3), (3, 2),
    ],
    [2e-1j]
):
    _coup_j = coup_j_base.copy()
    _coup_rk = coup_rk_base.copy()
    # _coup_rk = 10 * (np.random.rand(np.prod(model.shape), 1) - 0.5)
    # _coup_j[pidx(*target_plaquette, model.shape[1])] = pert_value
    # _coup_rk[pidx(*target_plaquette, model.shape[1])] = pert_value
    ham = build_model_and_H(_coup_j, _coup_rk, model.shape, basis, Site(*target_plaquette1), Site(*target_plaquette2), pert_value)
    fid = fidelity_vs_time(ham, scar_state, start, stop, num)
    ax.plot(
        tgrid,
        np.round(fid, 12),
        # marker="o",
        alpha=0.7,
        linestyle="--",
        # label=fr"$J_{{{target_plaquette}}} = ({pert_value.real:.1f}{pert_value.imag:+.3f}i)$",
        label=fr"$J_{{ \Box_{{{target_plaquette1}}}, \Box_{{{target_plaquette2}}} }} = ({pert_value.real:.2f}{pert_value.imag:+.2f}i)$",
    )


plt.rcParams['xtick.labelsize'] = 22
plt.rcParams['ytick.labelsize'] = 22
# ax.legend([f"epsilon = {epsilon}" for epsilon in [0.01, 0.05, 0.15]], fontsize=16, loc="upper right")
# plt.ylim(-0.05, 1.85)
# plt.title("QDM 4x4", fontsize=18)
plt.xlabel(r"$t$", fontsize=22)
plt.ylabel(r"$\mathcal{F}(t)$", fontsize=22)
plt.legend(
    fontsize=22,
    loc='lower left',
    framealpha=0.9
)
# plt.yscale('log')
plt.grid()
plt.tight_layout()
```
